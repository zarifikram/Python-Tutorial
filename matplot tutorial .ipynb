```python
%matplotlib inline
import matplotlib

import matplotlib.pyplot as plt
import random
import numpy as np
```

# figure : show the figure
# axis : work on the axis
# how to get these objects?


```python
fig, ax = plt.subplots()
X = np.linspace(0, 20, 10)
grad = -3
y_intercept = 10
Y = X*grad + y_intercept
```

```python
# plot function
plt.plot(X, Y)
plt.show()
```

```python
plt.xlabel("X VALUES")
plt.ylabel("Y VALUES")
plt.title("SINE CURVE")
plt.suptitle("A VISUALIZATION OF SINE GRAPH")
x = np.linspace(0, 2*np.pi, 20)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

# CREATE a random circle.( with random center and radius)
# plot it in matplot

# X = radius * cos ( theta )
# Y = radius * sin ( theta )

```python
radius = random.randint(10, 50)
theta = np.linspace(0, 2*np.pi, 30)

X = radius * np.cos(theta)
Y = radius * np.sin(theta)

fig, ax = plt.subplots()
ax.plot(X, Y)
plt.show()
```

```python
# two graphs in one plot

X = np.linspace(0, np.pi * 4, 40)
fig, ax = plt.subplots()
ax.plot(X, np.sin(X), X, np.cos(X), X, X + 1)
plt.show()
```

# two graphs in two rows


```python
fig, axes  = plt.subplots(2, 1)
```

# two graphs in two cols

```python
axes[0].plot(X, np.sin(X))
axes[1].plot(X, np.cos(X))
fig.show()
```
